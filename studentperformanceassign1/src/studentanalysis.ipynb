```python

import matplotlib.pyplot as plt
import pandas as pd

# Load the data from CSV file
df = pd.read_csv('../data_clean/student_data_cleaned.csv')

# Count the number of each gender
gender_counts = df['gender'].value_counts()

# Define a list of colors for each gender
colors = ["skyblue", "salmon"]

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the data from CSV file
df = pd.read_csv('../data_clean/student_data_cleaned.csv')

# Define a palette with different colors
palette = ["forestgreen", "indigo"]

# Set up the figure with subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# Box Plot: Scores by Test Preparation Course (Math)
sns.boxplot(data=df, x='test preparation course', y='math score', hue='test preparation course', palette=palette, ax=axes[0])
axes[0].set_title('Math Scores by Test Preparation Course')
axes[0].set_ylabel('Math Score')
axes[0].set_xlabel('Test Preparation Course')
axes[0].legend(title='Test Preparation Course')

# Box Plot: Scores by Test Preparation Course (Reading)
sns.boxplot(data=df, x='test preparation course', y='reading score', hue='test preparation course', palette=palette, ax=axes[1])
axes[1].set_title('Reading Scores by Test Preparation Course')
axes[1].set_ylabel('Reading Score')
axes[1].set_xlabel('Test Preparation Course')
axes[1].legend(title='Test Preparation Course')

# Box Plot: Scores by Test Preparation Course (Writing)
sns.boxplot(data=df, x='test preparation course', y='writing score', hue='test preparation course', palette=palette, ax=axes[2])
axes[2].set_title('Writing Scores by Test Preparation Course')
axes[2].set_ylabel('Writing Score')
axes[2].set_xlabel('Test Preparation Course')
axes[2].legend(title='Test Preparation Course')

# Adjust layout
plt.tight_layout()
plt.show()
```
